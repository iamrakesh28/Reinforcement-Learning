```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Consider a tank with inflow of x m^3 per second and outflow of y m^3 per second,  and the surface area of the tank is z  m^2 . Considering the initial level at t = 0 to be s0 , what is the capacity st over time?

```python
# SI units are used
class tank :
    def __init__(self, level, area):
        self.area = area
        self.volume = level * area
    def simulate(self, inflow, outflow, t):
        water = (inflow - outflow) * t * 1.0 + self.volume
        water = np.where(water < 0.0, 0.0, water)
        return water / self.area * 1.0

if __name__ == '__main__':
    # initial level = 5 m
    # surface area = 20 m^2
    # inflow = 2 m^3 / s
    # outflow = 4 m^3 / s
    A = tank(5, 20)
    xx = np.linspace(0, 100, 1000)
    yy = A.simulate(2, 5, xx)
    plt.plot(xx, yy)
    plt.xlabel('Time in seconds')
    plt.ylabel('Water level in meter')
    plt.title('Water Level vs Time')
    plt.show()
```

3. Consider a Queuing system where inter-arrival times and service times are exponentially distributed. In particular, customers arrive at times t1 , t2 , . . . , tn , . . ., and tn+1 − tn ∼ exp(λ). The time taken by server to serve nth customer is dn ∼ exp(μ).

```python
import queue

class system:
    def __init__(self, mu, lamda, n):
        self.mu = mu
        self.lamda = lamda
        self.n = n
        self.q = queue.Queue()
    def simulate(self, t):
        it = 1
        yy = [1]
        xx = [0]
        total = 0.0
        data = sample(self.mu, self.lamda)
        self.q.put(data)
        time = data[0]
        while it < t or self.q.qsize():
            if self.q.qsize():
                dn = self.q.get()[1]
                while dn >= time and it < t:
                    data = sample(self.mu, self.lamda)
                    it += 1
                    self.q.put(data)
                    dn = dn - time
                    total += time
                    yy.append(self.q.qsize() + 1)
                    xx.append(total)
                    time = data[0]
                total = total + dn
                yy.append(self.q.qsize())
                xx.append(total)
                if dn < time:
                    time = time - dn
            else:
                data = sample(self.mu, self.lamda)
                it += 1
                self.q.put(data)
                yy.append(self.q.qsize())
                total += time
                xx.append(total)
                time = data[0]
                
        return (xx, yy)
        
def sample(mu, lamda):
    delta_t = np.random.exponential(lamda)
    dn = np.random.exponential(mu)
    return (delta_t, dn)

if __name__ == '__main__':
    S1 = system(0.4, 0.6, 10)
    xx, yy = S1.simulate(S1.n)
    print(len(yy))
    plt.plot(xx, yy)
    plt.xlabel('Time in seconds')
    plt.ylabel('Customers in the queue')
    plt.title('No. of customers vs Time')
    plt.show()
    plt.step(xx, yy, where='post')
    plt.xlabel('Time in seconds')
    plt.ylabel('Customers in the queue')
    plt.title('No. of customers vs Time')
    plt.show()
```
